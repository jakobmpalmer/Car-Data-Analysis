```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
cars= pd.read_csv('/Users/jakepalmer/Desktop/DATASETS/Car_data.csv')
```

#1. Less than $45,000 and Convertable

```python
less_than_price = cars['MSRP'] < 45000
convert = cars['Vehicle Style'] == "Convertible"

cars.loc[(less_than_price & convert)]
```

```python
#2. 4 door Toyota w/ more than 25 mpg_H made after 2016
```

```python
four_door = cars['Number of Doors'] == 4.0
correct_mpg = cars['highway MPG'] > 25
year = cars['Year'] > 2016
toyota = cars['Make'] == 'Toyota'

cars.loc[(correct_mpg & four_door & year & toyota)]
```

```python
# 3. 6 cylinder Ford w/ less than 12 mpg_c
```

```python
six_cylinder = cars['Engine Cylinders'] == 6.0
ford = cars['Make'] == 'Ford'
our_mpgC = cars['city mpg'] < 12

cars.loc[(ford & our_mpgC & six_cylinder)]
```

```python
# 4. Highest mpg_H, under $30,000 w/ manuel transmission
price = cars['MSRP'] < 30000
transmission = cars['Transmission Type'] == 'MANUAL'

cars[(transmission & price)].sort_values(by=['highway MPG'], ascending=False)
```

```python
# 5. Highest HP in 2015 by Ford
year = cars['Year'] == 2015
ford = cars['Make'] == 'Ford'
horsepwr = 'Engine HP'
```

```python
cars.loc[(ford & year)].sort_values(by=['Engine HP'])
```
